# Data Cleaning

```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read th dataset from excel file and Print first 5 columns to get idea about it
df=pd.read_excel("Data/GSAF5-1.xls") #change path as per ur file location
df.head()
```

```python
#First subset 20 years of data
df_year = df[df.Year>1998.0]
df_year
```

```python
#Again Subsetting only USA data
USA_df = df_year[df_year["Country"]=="USA"]
USA_df
```

```python
#See the Structure of the dataset
USA_df.info()
```

```python
#We can see that the last two columns are useless so remove the last two columns
USA_df=USA_df.drop(USA_df.columns[[22,23]],axis=1)
```

```python
#Check for null value columns
USA_df.isnull().sum()
```

```python
#Filling the Activity with UNKNOWN values
USA_df["Activity"].fillna("UNKNOWN")

# Function to recode activities
def recode_act(a):
    if isinstance(a, str):
        a = a.lower()
        if 'surfing' in a:
            return "Surfing"
        elif 'swimming' in a:
            return "Swimming"
        elif 'fishing' in a:
            return "Fishing"
        elif 'bathing' in a:
            return "Bathing"
        elif 'wading' in a:
            return "Wading"
        elif 'diving' in a:
            return "Diving"
        elif 'standing' in a:
            return "Standing"
        elif 'snorkeling' in a:
            return "Snorkeling"
        else:
            return "Other"
    else:
        return "Other"
USA_df["Activity"] = USA_df["Activity"].apply(lambda x: recode_act(x))
```

```python
USA_df["Activity"].value_counts()
```

```python
USA_df["Fatal (Y/N)"].value_counts()
```

```python
#Filling the FATAL(Y/N) values with its mode value
USA_df["Fatal (Y/N)"].fillna('N',inplace=True)
```

```python
#Species column
USA_df["Species "] = USA_df["Species "].str.lower()
USA_df["Species "].value_counts()
```

```python
#recoding the species column
def code_species(a):
    if isinstance(a, str):
        a = a.lower()
        if 'white shark' in a:
            return "White Shark"
        elif 'shark involvement not confirmed' in a:
            return "Involvement unconfirmed"
        elif 'tiger  shark' in a:
            return "Tiger shark"
        elif 'bull shark' in a:
            return "Bull shark"
        elif 'no shark involvement' in a:
            return "No shark involvement"
        elif 'blacktip shark' in a:
            return "Blacktip shark"
        elif 'nurse shark' in a:
            return "Nurse shark"
        elif 'wobbegong shark' in a:
            return "Wobbegong shark"
        elif 'sandtiger shark' in a:
            return "Sandtiger shark"
        elif 'bronze whaler shark' in a:
            return "Bronze whaler shark"
        elif 'unidentified species' in a:
            return "Unidentified species"
        elif 'oceanic whitetip shark' in a:
            return "Oceanic whitetip shark"
        elif 'raggedtooth shark' in a:
            return "Raggedtooth shark"
        elif 'lemon shark' in a:
            return "Lemon shark"
        elif 'spinner shark' in a:
            return "Spinner shark"
        else:
            return "Other"
        
USA_df["Species "] = USA_df["Species "].apply(lambda x: code_species(x))
```

```python
USA_df["Species "].value_counts()
```

```python
#filling missing values in species
USA_df["Species "].isna().sum()
```

```python
#Replacing blank species values with unknown
USA_df["Species "].fillna("Unknown") 
```

```python
#finding mode in Sex and Age columns
print(USA_df['Sex '].mode())
print(USA_df['Age'].mode())
```

```python
#filling the Sex and Age values
USA_df['Sex '].fillna('M',inplace=True)
USA_df['Age'].fillna('15',inplace=True)
```

```python
#Printing cleaned dataset
USA_df
```

# Dataset Visualization

```python
#The below countplot gives the count of attacks with respect to state 
sns_Area=sns.countplot(x='Area',data=USA_df, palette="OrRd")
sns_Area.set_xticklabels(sns_Area.get_xticklabels(),rotation=40,ha="right")
plt.show()
```

```python
#Gives the count of different types of species
sns_Species=sns.countplot(x='Species ',data=USA_df, palette="OrRd")
sns_Species.set_xticklabels(sns_Species.get_xticklabels(),rotation=90,ha="right")
plt.show()
```

```python
#Below Scatter Plot describes the types of species attacked in a particular state
plt.scatter(USA_df['Species '],USA_df.Area,alpha=0.1)
plt.xticks(rotation=90)
```

```python
#Below Scatter Plot describes the types of attacks in a particular state
plt.scatter(USA_df.Area,USA_df.Type,alpha=0.1)
plt.xticks(rotation=90)
```

```python
#Box Plot for visualizing No.of Fatal Cases in 20 years
USA_df.boxplot(column='Year',by='Fatal (Y/N)')

#By observing the plot we can see that there are less number of fatal cases in the recent 10 years. 
```

```python
#The below crosstab graph displays the number of different types attacks on different types of species.
my_tab=pd.crosstab(index=USA_df['Species '],columns=USA_df['Type'])
my_tab.plot.bar()
```
